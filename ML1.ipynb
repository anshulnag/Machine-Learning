```python
def f(xx,yy):
    return np.sin(np.sqrt(xx**2+yy**2))
```

```python
xx=np.linspace(-1,5,10)
yy=np.linspace(-1,5,10)
X,Y=np.meshgrid(xx, yy)
Z=f(X,Y)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_wireframe(X,Y,Z,color='red')
```

```python
xx=np.linspace(-1,5,10)
yy=np.linspace(-1,5,10)
X,Y=np.meshgrid(xx, yy)
Z=f(X,Y)
fig=plt.figure()
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,cmap='magma_r',edgecolor='white')
```
