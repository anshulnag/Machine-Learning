```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from pytrends.request import TrendReq
trends=TrendReq()
```

```python
trends.build_payload(kw_list=["data analytics"])
data=trends.interest_by_region()
data=data.sort_values(by="data analytics", ascending=False)
data=data.head(10)
print(data)
```

```python
data.reset_index().plot(x="geoName",y="data analytics",figsize=(5,5),kind="bar")
plt.style.use('ggplot')
plt.show()
```

```python
data=TrendReq(hl='en-US',tz=360)
data.build_payload(kw_list=['data analytics'])
data=data.interest_over_time()
fig,ax=plt.subplots(figsize=(10,10))
data['data analytics'].plot()

plt.style.use('fivethirtyeight')
plt.show()
```
